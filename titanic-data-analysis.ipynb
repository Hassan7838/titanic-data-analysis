```python
# Expolatory Data Analysis on Titanic Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
print("\n=== First 5 Rows===")
print(titanic.head())

print("\n===Datatypes===")
print(titanic.dtypes)

print("\n===Summary stats===")
print(titanic.describe())

print("\n===Missing Values===")
print(titanic.isnull().sum())

print("\n===Unique Counts===")
for column in titanic.columns:
  print(f"{column}: {titanic[column].nunique()} unique values")
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.countplot(x='sex',data=titanic)
plt.title('Passengers gender distribution')

plt.subplot(2,2,2)
sns.countplot(x='class',data=titanic)
plt.title('Passengers Class distribution')

plt.subplot(2,2,3)
sns.histplot(titanic['age'].dropna(),bins=20)
plt.title('Age distribution')

plt.subplot(2,2,4)
sns.scatterplot(x='age',y='fare',data=titanic)
plt.title('Age vs Fare paid')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
numerical_cols = titanic.select_dtypes(include=['int64','float64']).columns
sns.heatmap(titanic[numerical_cols].corr(), annot=True , cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
